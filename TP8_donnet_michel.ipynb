(c) We write

$$f_N (t) = \frac{1}{2} a_0 + \sum_{k=1}^N a_k \cos(k \omega_0 t) + b_k \sin(k \omega_0 t).$$

Implement the functions $\tilde{f}(t)$ and $f_N(t)$ in Python and plot them on the same graph for $t \in [-5, 5]$.
Do several plots for $N = 0, 1, 3, 5, 10, 100$.

```python
import numpy as np
import matplotlib.pyplot as plt



def f_tilda(t: np.ndarray) -> np.ndarray:
    y = t.copy()
    mask_1 = ((y % 4 > 1) & (y % 4 <= 3)).astype(int)
    mask_2 = mask_1 - 1
    return - mask_1 - mask_2

def f(t, N):
    if N < 1:
        return np.zeros(len(t))
    result = 0
    for k in range(1, N + 1):
        result += 4 / (k * np.pi) * np.sin(k * np.pi / 2) * np.cos(k * np.pi/2 * t)
    return result

N = [0, 1, 3, 5, 10, 100]

x = np.linspace(-5, 5, 201)


for i in N:
    plt.figure()
    plt.title(f"N = {i}")
    plt.plot(x, f_tilda(x), label="f_tilda")
    plt.plot(x, f(x, i), label=f"$f{i}$")

    plt.legend()
    plt.show()
```

(b) Fix a real number $T \geq 2$, and consider $\left.h\right|_{[-T/2, T/2]}$. Define $h_T$ as the periodic extension of this function on all $\mathbb{R}$.

On separate graphs, visualize $h(t)$, $h_2(t)$, $h_4(t)$ and $h_8(t)$ for $t \in [-10, 10]$.

```python
import matplotlib.pyplot as plt
import numpy as np

def h(x: np.ndarray) -> np.ndarray:
    y = x.copy()
    mask = (y >= -1) & (y <= 1)
    y[mask] = 1 - np.abs(y[mask])
    y[~mask] = 0
    return y

def delta(x: np.ndarray, T: float) -> np.ndarray:
    return x % T == 0

x = np.linspace(-10, 10, 201)

T = [2, 4, 8]
index = 2

plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.title("h(t)")
plt.plot(x, h(x))

for i in T:
    plt.subplot(2, 2, index)
    plt.title(f"$h_{i}(t)$")
    plt.plot(x, np.convolve(h(x), delta(x, i), mode='same'))
    index += 1
plt.show()
```

```python
def H(w: np.ndarray) -> np.ndarray:
    return 2 / w**2 * (1 - np.cos(w))


def points(T: int) -> np.ndarray:
    x = np.array([k for k in range(-3 * T, 3 * T)]).astype(float)
    y = []
    for i in x:
        if i != 0:
            y.append(T/(2 * i**2 * np.pi**2) * (1 - np.cos(i * (2 *np.pi) / T)))
        else:
            y.append(0)
    y = np.array(y) * T
    x *= 2 * np.pi/T
    return x, y
        
    
T = [2, 4, 8, 50, 100]
t = np.linspace(-20, 20, 200)

for i in T:
    points_x, points_y = points(i)
    plt.figure()
    plt.title(f"T = {i}")
    plt.plot(t, H(t))
    plt.plot(points_x, points_y, 'o')
    plt.show()
```

Using complex analysis, one can show that that the Fourier transform of a Gaussian function is again a Gaussian function. Namely :

$$
g(t) = e^{- \alpha t^2} \implies G(w) = \sqrt{\frac{\pi}{\alpha}} \cdot e^{- \frac{w^2}{4 \alpha}}

```python
def g(w: np.ndarray, alpha: float) -> np.ndarray:
    return np.exp(-alpha * w**2)

def G(w: np.ndarray, alpha: float) -> np.ndarray:
    return np.sqrt(np.pi / alpha) * np.exp(- w**2 / 4 * alpha)

alpha = [0.1, 1, 2, 10]
x = np.linspace(-10, 10, 201)


for i in alpha:
    plt.figure()
    
    plt.subplot(121)
    plt.title(f"g(x) with alpha={i}")
    plt.plot(x, g(x, i))

    plt.subplot(122)
    plt.title(f"G(x) with alpha={i}")
    plt.plot(x, G(x, i))

    plt.show()
```

(c) Write a python function *represent_complex_signal*$(x,y)$, where

- y is a complex numpy array defined by $y = f(x)$.

The function should plot four different representations of the pairs $(x, y)$. Namely :


- Real part : using the pair $(x, $ *np.real*$(y))$
- Imaginary part : using the pair $(x, $ *np.imag*$(y))$
- Magnitude : using the pair $(x, $ *np.abs*$(y))$
- Phase : using the pair $(x, $ *np.angle*$(y))$

```python
def represent_complex_signal(x,y, title = "G(w)"):
    plt.figure(figsize=(10, 10))
    
    plt.subplot(221)
    plt.title(f"Real part of {title}")
    plt.plot(x, np.real(y))

    plt.subplot(222)
    plt.title(f"Imaginary part of {title}")
    plt.plot(x, np.imag(y))

    plt.subplot(223)
    plt.title(f"Magnitude of {title}")
    plt.plot(x, np.abs(y))

    plt.subplot(224)
    plt.title(f"Phase of {title}")
    plt.plot(x, np.angle(y))
```
