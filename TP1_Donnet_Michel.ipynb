(c) Write your own function that adds salt and pepper noise to the image. As inputs the function should take: input image - np.ndarray, p - percentage of the pixels to be replaced with noise. Make sure that your function works with singlechannel and multichannel images

```python
import random

def add_salt_and_pepper(im_input: np.ndarray, p: float):
    # If the image has three channels, we convert it to grayscale...
    if im_input.shape[2] == 3:
        img = sk.color.rgb2gray(im_input)
    else:
        img = im_input
    rows, columns = img.shape
    # We create an matrix with a percentage of 0 of p, and then we multiply the two matrix for have a matrix with p percent of 0
    random = np.uint8(np.random.binomial(1, 1 - p, (rows, columns)))
    img = img * random

    # Then, we create the logical invert of the random matrix (so with p percent of 1)
    invert_random = np.uint8(np.ones([rows, columns]))
    invert_random = np.bitwise_xor(random, invert_random)

    # Finally we multiply this new matrix with a random matrix, and we add it to the image (that works because we just put a value where we created zeros before) 
    invert_random = np.float32(invert_random * np.random.rand(rows, columns))
    img = img + invert_random
    
    # We print the result
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(f'S&P with p = {p}')
    plt.show()
    
add_salt_and_pepper(lena, 0.25)        
```
